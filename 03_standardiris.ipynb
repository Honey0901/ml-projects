```python
#importing
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
df=pd.read_csv("/content/Iris.csv")
df.sample(5)
```

```python
df.isnull().sum()
```

```python
df.shape
df.info()
```

```python
x = df.drop('Species',axis=1)
y = df['Species']
```

```python
#train test split
x_train ,x_test  ,y_train ,y_test  =train_test_split(x.iloc[:,1:],y,random_state=42,test_size=0.3)
```

```python
#scalling
sc=StandardScaler()
sc.fit(x_train)
```

```python
sc_trans=sc.fit_transform(x_train)
sc_trans = pd.DataFrame(sc_trans,columns=x_train.columns)
sc_trans.head()
```

```python
sc_test=sc.fit_transform(x_test)
sc_test = pd.DataFrame(sc_test,columns=x_test.columns)
sc_test.head(5)
```

```python
x_test.head()
```

```python
#model fitting
model=LogisticRegression()
model.fit(sc_trans,y_train)
```

```python
y_pred=model.predict(sc_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
