```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv("/content/matches.csv")
df1=pd.read_csv("/content/most_runs_average_strikerate.csv")
```

```python
#cleaning data
df.shape
```

```python
df.isnull().sum()
df.dropna()
df.shape
```

```python
df1.isnull().sum()
df1.dropna()
df1.shape
```

```python
#most successfull batsman
df['winner'].value_counts().nlargest(5)
```

```python
#toss winner vs match winner
df['top_vs_match']=df['toss_winner']==df['winner']
toss_match = df['top_vs_match'].value_counts()

print(toss_match)
plt.bar(toss_match.index,toss_match.values)
```

```python
#top batsman
tp=df1.groupby('batsman')['total_runs'].sum().sort_values(ascending=False).nlargest(5)
print(tp)
```

```python
win=df['winner'].value_counts()
plt.barh(win.index,win.values)
```

```python
toss_dec=df['toss_decision'].value_counts()
# print(toss_dec)
plt.pie(toss_dec.values,labels=toss_dec.index,autopct='%1.1f%%')
```
