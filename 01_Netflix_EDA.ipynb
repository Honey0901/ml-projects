## 🔍 Why Use `df.shape`?

### ✅ 1. **What it does:**

* This line **shows the size of your dataset**.
* It returns a **tuple**: `(number_of_rows, number_of_columns)`

For example:

```python
df.shape
```

Might return:

```
(8807, 12)
```

Which means:

* **8807 rows** → 8807 shows/movies
* **12 columns** → 12 pieces of information per entry (like title, country, release year, etc.)

---

### 🎯 2. **Why it’s useful:**

* Helps you **understand how big your data is** before working with it.
* You can quickly tell:

  * Do I have enough data?
  * Is there too much to handle in memory?
  * Do I need to clean or reduce anything?
* It helps **debug issues** — for example, if you accidentally load the wrong file and it returns (0, 0), you instantly know something went wrong.

---

### ⚠️ 3. **What happens if you skip it?**

Nothing will **break**, but:

* You’ll be working **blind**.
* You won’t know how many records you're dealing with.
* You could make mistakes later when plotting, grouping, or filtering.
* If the file didn't load correctly, you wouldn't know early.

---

### ✅ Best Practice:

Always do `df.shape` and `df.head()` after loading your data — they are **safe, fast, and informative**.

## 🧩 **Step 2 Code: Understanding the DataFrame**

```python
df.info()
```

---

### 🔹 **1. `df.info()`**

#### ✅ What it does:

* Prints a summary of the DataFrame:

  * Total rows (entries)
  * Each column’s name, number of non-null values, and data type (int, object, etc.)

#### 🎯 Why it’s useful:

* Helps you:

  * Understand the overall structure of the dataset
  * See **which columns have missing data**
  * Identify if **any data types need conversion**, like converting dates from object to datetime

#### ⚠️ What happens if you skip it:

* You might miss:

  * Hidden missing values (nulls)
  * Wrong data types (like strings for numbers)
* That can break future analysis or give incorrect results

#### 🧠 Best Practices:

* Always run `df.info()` immediately after `df.read_csv()` and `df.head()`
* Use it again after cleaning to check if nulls are gone or types are updated

---

### 🔹 **2. `df.isnull().sum()`**

#### ✅ What it does:

* Checks **each column** for how many **missing (`NaN`) values** it has.
* Returns a count for every column.


#### 🎯 Why it’s useful:

* Tells you **which columns need cleaning or filling**
* Helps you decide if a column is:

  * Fine to keep (few missing values)
  * Needs imputation (fill in missing)
  * Should be dropped (too many missing)

#### ⚠️ What happens if you skip it:

* You may run into errors when:

  * Plotting charts
  * Using `.groupby()` or `.sort_values()`
  * Training ML models later
* Missing values can silently **ruin your analysis**

#### 🧠 Best Practices:

* Always run this after `df.info()`
* If any column has **a lot of missing data**, plan to:

  * Drop the column
  * Fill missing data (with mean, median, or "Unknown", etc.)
* Combine this with visual null checks (e.g. using `sns.heatmap(df.isnull())` later)

---

### ✅ Summary of Step 2:

| Command             | Purpose                               |
| ------------------- | ------------------------------------- |
| `df.info()`         | Overview of data types + missing data |
| `df.isnull().sum()` | Exact number of missing values        |

These two together give you a full picture of the **health and structure** of your dataset before you touch or analyze anything else.

````python
country_count = df['country'].value_counts().head(10)
print(country_count)

sns.barplot(x=country_count.values , y=country_count.index)
plt.title("Top 10 Countries")
plt.show()
````

````python
# 'listed_in' column contains genres (comma-separated strings)
#aa baki che

# Split genres and flatten the list
from collections import Counter

genres = df['listed_in'].dropna().apply(lambda x: x.split(', '))
flat_genres = [genre for sublist in genres for genre in sublist]

top_genres = Counter(flat_genres).most_common(10)

# Convert to DataFrame for plotting
genres_df = pd.DataFrame(top_genres, columns=["Genre", "Count"])

sns.barplot(data=genres_df, x="Count", y="Genre")
plt.title("Top 10 Genres on Netflix")
plt.show()
````
